```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.figure(figsize=(11, 8))
plt.hist(data, bins=30, alpha=0.7, color='red', edgecolor='black')

plt.title('Histograma de Datos Aleatorios')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.grid(True)
plt.show()
```
