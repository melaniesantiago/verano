```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
datos = [12, 15, 14, 10, 8, 11, 13, 15, 14, 10, 9, 12, 13, 14, 15]

# Crear el histograma
frecuencias, bins = np.histogram(datos, bins=5)

# Calcular los puntos medios de los intervalos
puntos_medios = 0.5 * (bins[1:] + bins[:-1])

# Graficar el polígono de frecuencias
plt.plot(puntos_medios, frecuencias, marker='o', linestyle='-', color='blue', label='Polígono de frecuencias')

# Opcional: Añadir el histograma como referencia
plt.hist(datos, bins=5, alpha=0.3, color='gray', label='Histograma')

# Personalización del gráfico
plt.title('Polígono de Frecuencias')
plt.xlabel('Intervalos')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```
