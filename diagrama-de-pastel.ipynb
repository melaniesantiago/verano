```python
import matplotlib.pyplot as plt

# Datos de ejemplo
etiquetas = ['Manzanas', 'Plátanos', 'Cerezas', 'Uvas']
valores = [15, 30, 45, 10]

# Crear gráfico de pastel
plt.figure(figsize=(6, 6))
plt.pie(valores, labels=etiquetas, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Frutas')
plt.axis('equal')  # Para que el pastel sea un círculo
plt.show()
```
